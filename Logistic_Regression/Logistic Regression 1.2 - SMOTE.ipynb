```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from collections import Counter
from imblearn.under_sampling import RandomUnderSampler

df = pd.read_excel('NewEncodedData.xlsx')
df = df.drop('Unnamed: 0', axis = 1)
df.head()
```

```python
# Define X and y correctly
y = df['y']
X = df.drop('y', axis=1)

# Summarize the class distribution
print("Original class distribution:", Counter(y))

# Initialize and apply SMOTE
oversample = SMOTE()
X_smote, y_smote = oversample.fit_resample(X, y)

# Summarize the new class distribution
print("New class distribution after SMOTE:", Counter(y_smote))

# Create a new DataFrame with the oversampled data
df_smote = pd.DataFrame(X_smote, columns=X.columns)
df_smote['y'] = y_smote  # Add the target variable back

df_smote
```

```python
df_smote.to_excel("SMOTE_oversampled_data.xlsx", index=False)
```

Undersampling

```python
df_new = pd.read_excel('Bank Marketing Data_raw.xlsx')
# df_new = df.drop('Unnamed: 0', axis = 1)
df_new.head()
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTEENN


y = df_new['y']
X = df_new.drop('y', axis=1)

# Apply Random Undersampling to the oversampled dataset
undersample = RandomUnderSampler()
X_balanced, y_balanced = undersample.fit_resample(X, y)
print("After Undersampling:", Counter(y_balanced))

# Create a new DataFrame with the balanced data
df_balanced = pd.DataFrame(X_balanced, columns=X.columns)
df_balanced['y'] = y_balanced
```
